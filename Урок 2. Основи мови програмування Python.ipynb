## Частина 1. Встроєні типи даних

    Навіщо потрібні вбудовані типи даних (об'єкти)?  Вбудовані об'єкти спрощують створення програм. Оскільки позбавляють від необхідності реалізації структур даних. 
    
    Вбудовані об'єкти – це компоненти розширення. Для вирішення складних задач ви можете створювати власні об'єкти, використовуючи для цього вбудовані класи мови Python.
    
    Вбудовані об'єкти часто ефективніші, ніж створені вручну структури даних. Вбудовані типи мови програмування Python використовують вже оптимізовані структури даних, реалізовані на мові програмування Сдля досягнення високої продуктивності.
    
    Влаштовані об'єкти - це стандартна частина мови програмування. В деякій мірі Python багато що запозичує як з мов, що покладаються на використання вбудованих інструментальних засобів (таких як LISP), так і що покладаються на майстерність програміста, який повинен виконати власну реалізацію інструментів і структур даних (таких як C++). У мові Python можна створювати власні типи об'єктів, але на самому початку робити це не рекомендується.
    
    Більш того, через те, що вбудовані компоненти є стандартними складовими мови Python, вони завжди залишаються незмінними, тоді як власні структури мають властивість змінюватися від випадку до випадку. Ми розглянемо наступні вбудовані типи даних:
    
    Числа (7382, 3.14, 3+4j, Decimal, Fraction)
    Рядки ('net', "your's", u'радость')
    Списки ([1, [2, 'three'], 4])
    Словники({'Alex': 2, 'Brian': 4})
    Кортежі ('Leo', 21.7, 'single')
    Множини (set(1,2,3), {'a', 'b', 'c'})
    Файли (open('myfile', 'r'))
    
   ## Динамічна типізація
   
    Типи даних в мові Python визначаються автоматично під час виконання, а не в результаті оголошення у програмному коді. Змінні створюються при виконанні операції присвоєння, можуть посилатися на об'єкти будь-яких типів і їм мають бути присвоєні деякі значення, перш ніж до них можна буде звернутися.
    Змінні - це записи у системній таблиці, де передбачено місце для зберігання посилань на об'єкти.
    Об'єкти - це область пам'яті з об'ємом, достатнім для представлення значеннь цих об'єктів. Кожний об'єкт має два стандартні поля: описувач типу, який використовується  для зберігання інформації про тип об'єкта, і лічильник посилань, використаний для визначення моменту, коли пам'ять, займана об'єктом, може бути звільнена.
    Посилання - це покажчики на об'єкти.
